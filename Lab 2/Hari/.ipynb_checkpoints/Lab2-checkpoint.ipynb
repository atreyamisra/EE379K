# Problem 4 Part 2 Write a program that on input Name returns the frequency for men and women of the name Name

```python
def part2(Name):
    f_count = 0
    m_count = 0
    for year in range(1880, 2015 + 1):
        data = pd.read_csv('Names\yob' + str(year) + '.txt', header=None)
        df = data.loc[data[0] == Name]
        if (df.iloc[0, 1]) == 'M':
            m_count += df.iloc[0, 2]
        else:
            f_count += df.iloc[0, 2]

        if len(df.index) > 1:
            if (df.iloc[1, 1]) == 'M':
                m_count += df.iloc[1, 2]
            else:
                f_count += df.iloc[1, 2]

    print 'Females: ' + str(f_count) + '\nMales: ' + str(m_count)


part2('Mary')
```

# Problem 3 Part 3

```python
import pandas as pd

def part3(Name):
    count = 0
    total = 0
    for year in range(1880, 2015 + 1):
        data = pd.read_csv('Names\yob' + str(year) + '.txt', header=None)
        df = data.loc[data[0] == Name]
        count += df.iloc[0, 2]
        total += sum(data[2].tolist())
        #print total

        if len(df.index) > 1:
            count += df.iloc[1, 2]

    print 'Relative Frequency: ' + str(count*100.0/total) + '%'


part3('John')
```
