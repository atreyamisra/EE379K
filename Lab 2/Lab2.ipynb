```python
import numpy as np
import matplotlib.pyplot as plt

beta = -3

betas = []
errors = []

for i in range(0, 1000):
   e = np.random.normal(size=150)
   y = e + beta

   x = np.random.normal(size=150)
   errors.append(x.T.dot(e) / x.T.dot(x))
   betas.append(x.T.dot(y) / x.T.dot(x))

print 'Empirical standard deviation of errors: ' + str(np.std(errors))
n = list(map(lambda x: 1 / np.sqrt(x), range(10, 10010, 100)))
n = list(n) # convert to list

stderrors = []
errors = []
for i in range(10, 10010, 100):
   for j in range(0, 100):
       x = np.random.normal(size=i)
       e = np.random.normal(size=i)
       y = e + beta
       errors.append(x.T.dot(e) / x.T.dot(x))
   stderrors.append(np.std(errors))

plt.scatter(range(10, 10010, 100), stderrors)
plt.plot(range(10, 10010, 100), n)
plt.show()
```
